```python
import osmnx as ox
import networkx as nx

# 1. Define the combined area (bounding box or union of places)
places = ["Lombok, Indonesia", "Bali, Indonesia"]

# 2. Get the network graph for both regions in one graph
G = ox.graph.graph_from_place(places, network_type="drive")

# 3. Define weight for routing
weight = "length"

# 4. Select origin and destination nodes
# You can use nearest nodes based on lat/lon coordinates
orig_point = (-8.6500, 116.3249)  # Example: Mataram, Lombok
dest_point = (-8.4095, 115.1889)  # Example: Denpasar, Bali

orig_node = ox.distance.nearest_nodes(G, X=orig_point[1], Y=orig_point[0])
dest_node = ox.distance.nearest_nodes(G, X=dest_point[1], Y=dest_point[0])

# 5. Get the shortest route between nodes
route = nx.shortest_path(G, orig_node, dest_node, weight=weight)

# 6. Convert the route to a GeoDataFrame
route_edges = ox.routing.route_to_gdf(G, route)

# 7. Explore the route on an interactive map
m = route_edges.explore(
    tiles="cartodbpositron",
    style_kwds={"weight": 5, "color": "blue"}
)

# 8. Optional: Save the map as an HTML file
m.save("route_map.html")

# Display in notebook (if applicable)
m
```
